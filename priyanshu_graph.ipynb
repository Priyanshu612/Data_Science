```python
import matplotlib.pyplot as plt
```

```python
plt.plot(reshaped_df.index,reshaped_df["java"])
```

```python
plt.figure(figsize=(16,8))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(reshaped_df.index,reshaped_df["java"],label="java")
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.ylabel("no of posts",fontsize=14)
plt.ylim(0,35000)
plt.plot(reshaped_df.index,reshaped_df["java"],label="java")
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.ylabel("no of posts",fontsize=14)
plt.ylim(0,35000)
plt.plot(reshaped_df.index,reshaped_df["java"],label="java")
plt.plot(reshaped_df.index,reshaped_df["python"],label="python")
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.ylabel("no of posts",fontsize=14)
plt.ylim(0,35000)
for colum in reshaped_df.columns:
    plt.plot(reshaped_df.index,reshaped_df[colum],label=colum,linewidth=3)
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.ylabel("no of posts",fontsize=14)
plt.ylim(0,35000)
for colum in reshaped_df.columns:
  plt.plot(reshaped_df.index,reshaped_df[colum],label=colum,linewidth=3)
plt.legend(fontsize=16)
plt.show()

```

```python

rolling_df = reshaped_df.rolling(window=6).mean()
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.ylabel("no of posts",fontsize=14)
plt.ylim(0,35000)
for colum in rolling_df.columns:
  plt.plot(rolling_df.index,rolling_df[colum],label=colum,linewidth=3)
plt.legend(fontsize=16)
plt.show()
```
