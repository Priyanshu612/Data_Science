```python
import cv2
import os
from google.colab.patches import cv2_imshow # Keep this import

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the image
# Use cv2.imread for reading a static image file
imp_img = cv2.imread("elon.jpeg")

# Check if the image was loaded successfully
if imp_img is None:
    print("Error: Could not read the image file.")
    # You might want to exit or handle this error appropriately
else:
    # Convert to grayscale
    gray = cv2.cvtColor(imp_img, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

    # Draw rectangles around the faces
    # Iterate over the faces found
    for (x, y, w, h) in faces:
        cv2.rectangle(imp_img, (x, y), (x + w, y + h), (255, 255, 0), 2)

    # Display the image using cv2_imshow for Colab compatibility
    # Replace cv2.imshow with cv2_imshow
    cv2_imshow(imp_img)

    # Note: cv2_imshow does not require cv2.waitKey(0) or cv2.destroyAllWindows()
    # as it displays the image directly in the output cell.
    # You can remove these lines.
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()


# Note: Since we used cv2.imread, there is no VideoCapture object to release.
# The original imp_img variable is now the image data itself.
```
